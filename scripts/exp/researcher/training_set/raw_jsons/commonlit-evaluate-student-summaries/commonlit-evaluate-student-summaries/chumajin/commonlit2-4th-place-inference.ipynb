# Submission

```python
import pandas as pd 
import numpy as np 
import os
```

```python
targets = ["content","wording"]
```

# Submission

```python
def make_finpreds(exp1,exp2):
    
    sample = pd.read_csv(f"/kaggle/working/submission{exp1}.csv")
    sample2 = pd.read_csv(f"/kaggle/working/submission{exp2}.csv")
    samples = [sample,sample2]
    finpreds = np.mean([ddf[targets].values for ddf in samples],axis=0)
    return finpreds
```

```python
def make_finpreds2(exp1):
    
    sample = pd.read_csv(f"/kaggle/working/submission{exp1}.csv")
    return sample[targets].values
```

```python
finpreds1 = make_finpreds2(91)
finpreds2 = make_finpreds2(22)
finpreds3 = make_finpreds2(63)
finpreds4 = make_finpreds2(72)
finpreds5 = make_finpreds(2,3)
finpreds6 = make_finpreds2(259)
finpreds7 = make_finpreds2(331)
```

```python
samples = [finpreds1,finpreds2,finpreds3,finpreds4,finpreds5,finpreds6, finpreds7]

finpreds = np.mean(samples,axis=0)
```

```python
sample = pd.read_csv("/kaggle/input/commonlit-evaluate-student-summaries/sample_submission.csv")
```

```python
sample["content"] = finpreds[:,0]
sample["wording"] = finpreds[:,1]
sample
```
