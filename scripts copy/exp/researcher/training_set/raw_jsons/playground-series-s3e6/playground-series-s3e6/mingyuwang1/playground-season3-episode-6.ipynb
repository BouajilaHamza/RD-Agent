```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from xgboost import XGBRegressor
```

```python
df_sub=pd.read_csv("/kaggle/input/playground-series-s3e6/sample_submission.csv")
df_train=pd.read_csv("/kaggle/input/playground-series-s3e6/train.csv")
df_test=pd.read_csv("/kaggle/input/playground-series-s3e6/test.csv")
```

```python
df_train.drop(['id'], axis=1, inplace=True)
```

```python

df_train.drop(['cityCode'], axis=1, inplace=True)
```

```python
df_train.dropna(inplace=True)
```

```python
X=df_train.iloc[:,:15]
y=df_train.price 
```

```python
# Splitting
train_X, test_X, train_y, test_y = train_test_split(X, y,test_size = 0.3, random_state =1234)
```

```python
model=XGBRegressor(max_depth=3, learning_rate=0.24 , n_estimators=50000, objective='reg:linear', booster='gbtree')
XGB=model.fit(train_X,train_y)
prediction=XGB.predict(test_X)
```

```python
df_test.drop(['id','cityCode'], axis=1, inplace=True)
```

```python
test_pred = XGB.predict(df_test)
```

```python
df_lb_score = pd.read_csv('/kaggle/input/worstscoreagainpa/submissionnewbestscore.csv')
```

```python
df_lb_score.head()
```

```python
df_sub.price=test_pred*0.2+df_lb_score['price']*0.8
```

```python
df_sub.to_csv('submission.csv', index = False)
pd.read_csv('submission.csv')
```
